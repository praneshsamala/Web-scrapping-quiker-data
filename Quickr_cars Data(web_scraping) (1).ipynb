```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.quikr.com/cars/used+Maruti-Suzuki+Swift+cars+all-india+z1399vbd"
```

```python
headers={"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5)AppleWebKit/605.1.15 (KHTML, like Gecko)Version/12.1.1 Safari/605.1.15"}
```

```python
res=requests.get(url,headers=headers)
```

```python
res
```

```python
soup=BeautifulSoup(res.text,"lxml")
```

```python
div=soup.find_all("div",class_="qc-ads")
```

```python
car=soup.find_all("a",class_="qc-ads__card")
```

```python
len(car)
```

```python
name=[]
prime=[]
price=[]
```

```python
for i in car:
    try:
        name.append(i.find("h2").text)
    except:
        np.nan
    try:
        prime.append(i.find("div",class_="prime-features").text)
    except:
        np.nan
    try:
        price.append(i.find("div",class_="price").text)
    except:
        np.nan
```

```python
div=soup.find_all("div")
```

```python
df=pd.DataFrame({"name":name,"prime":prime,"price":price})
```

```python
models=["Wagon-R","i20","Duster","Swift-Dezire","Jeep","i10","City","Ertiga","Kwid","Triber","Creta","Thar","Innova","Fortuner","Glanza","Verna","Scorpio","Scorpio_n","Kodiaq","Rapid","Kiger","Nexon","Hexa","Tiago"]
```

```python
brand=["Maruti-Suzuki","Hyundai","Renault","Maruti-Suzuki","Mahindra","Hyundai","Honda","Maruti-Suzuki","Renault","Renault","Hyundai","Mahindra","Toyota","Toyota","Toyota","Hyundai","Mahindra","Mahindra","Skoda","Skoda","Renault","Tata","Tata","Tata"]
```

```python
car_list=dict(zip(models,brand))
```

```python
car_list
```

```python
type(car_list)
```

```python
for k,v in car_list.items():
    url="https://www.quikr.com/cars/used+{}+{}+cars+all-india+z1399vbd".format(v,k)
    headers={"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5)AppleWebKit/605.1.15 (KHTML, like Gecko)Version/12.1.1 Safari/605.1.15"}
    res=requests.get(url,headers=headers)
    soup=BeautifulSoup(res.text,"lxml")
    car=soup.find_all("a",class_="qc-ads__card")
    name=[]
    prime=[]
    price=[]
    for i in car:
        try:
            name.append(i.find("h2").text)
        except:
            np.nan
        try:
            prime.append(i.find("div",class_="prime-features").text)
        except:
            np.nan
        try:
            price.append(i.find("div",class_="price").text)
        except:
            np.nan
    cars={"name":name,"prime":prime,"price":price}
    df1=pd.DataFrame(cars)
    frame=[df,df1]
    df=pd.concat(frame)
    print(df)
```
